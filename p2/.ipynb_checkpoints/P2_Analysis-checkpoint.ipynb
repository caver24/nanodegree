```python
#inport statments
import pandas as pd
%pylab inline
import matplotlib.pyplot as plt
import seaborn as sns

#import data in to DF
titanic_data_df = pd.read_csv('titanic_data.csv')

titanic_data_df #get a feel
```

#  Gender Survival

```python
#plot some stats
titanic_data_grouped_by_sex = titanic_data_df.groupby('Sex')

titanic_data_grouped_by_sex.count()['PassengerId'].plot.pie()

survivor_data_by_sex = pd.DataFrame({})
survivor_data_by_sex['not_survived'] = titanic_data_grouped_by_sex.count()['Survived'] - titanic_data_grouped_by_sex.sum()['Survived']
survivor_data_by_sex = survivor_data_by_sex.join(titanic_data_grouped_by_sex.sum()['Survived'])

print survivor_data_by_sex

#plot the ratios of suvived vs not survived
survivor_data_by_sex.plot.pie(subplots=True)
```

## Class Survival

```python
#plot some stats
titanic_data_grouped_by_class = titanic_data_df.groupby('Pclass')

titanic_data_grouped_by_class.count()['PassengerId'].plot.pie()

survivor_data_by_class = pd.DataFrame({})
survivor_data_by_class['not_survived'] = titanic_data_grouped_by_class.count()['Survived'] - titanic_data_grouped_by_class.sum()['Survived']
survivor_data_by_class = survivor_data_by_class.join(titanic_data_grouped_by_class.sum()['Survived'])

print survivor_data_by_class

#plot the ratios of suvived vs not survived
survivor_data_by_class.plot.pie(subplots=True)
```
